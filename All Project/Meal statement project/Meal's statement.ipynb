```python

"""
Meal Cost Calculator
A program to calculate shared meal costs among roommates/friends
"""

def get_person_data(name):
    """Get meal and deposit data for a person."""
    print(f"\n--- {name.upper()} ---")
    meals = int(input(f"Input meals for {name}: "))
    deposit = int(input(f"Input deposit for {name} (Tk): "))
    return meals, deposit

def calculate_balance(meals, deposit, meal_rate):
    """Calculate balance for a person."""
    return deposit - (meals * meal_rate)

def display_result(name, balance):
    """Display result for a person."""
    if balance > 0:
        print(f"{name} will collect money: Tk {balance:.2f}")
    else:
        print(f"{name} has to deposit money: Tk {abs(balance):.2f}")

def main():
    """Main function to run the meal cost calculator."""
    # List of people
    people = ["Tasib", "Rashed", "Sajib", "Emon", "Yousuf", "Wany", "Atif", "Akib"]

    # Dictionary to store data
    data = {}

    print("=" * 50)
    print("MEAL COST CALCULATOR")
    print("=" * 50)

    # Collect data for all people
    for person in people:
        meals, deposit = get_person_data(person)
        data[person] = {"meals": meals, "deposit": deposit}

    # Calculate totals
    total_meals = sum(person_data["meals"] for person_data in data.values())
    total_deposit = sum(person_data["deposit"] for person_data in data.values())

    # Calculate meal rate
    if total_meals == 0:
        print("\nError: Total meals cannot be zero!")
        return

    meal_rate = total_deposit / total_meals

    # Display summary
    print("\n" + "=" * 50)
    print("SUMMARY")
    print("=" * 50)
    print(f"Total Meals: {total_meals}")
    print(f"Total Deposit: Tk {total_deposit}")
    print(f"Meal Rate: Tk {meal_rate:.2f} per meal")

    # Calculate and display results
    print("\n" + "=" * 50)
    print("INDIVIDUAL RESULTS")
    print("=" * 50)

    for person, person_data in data.items():
        balance = calculate_balance(person_data["meals"], person_data["deposit"], meal_rate)
        display_result(person, balance)

if __name__ == "__main__":
    main()
```

```python

"""
Meal Cost Calculator - GUI Version using Tkinter
A graphical interface for calculating shared meal costs
"""

import tkinter as tk
from tkinter import ttk, messagebox

class MealCalculatorGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Meal Cost Calculator")
        self.root.geometry("700x650")
        self.root.resizable(False, False)

        # Configure colors
        self.bg_color = "#f0f4f8"
        self.frame_color = "#ffffff"
        self.button_color = "#4CAF50"
        self.header_color = "#2196F3"

        self.root.configure(bg=self.bg_color)

        # People list
        self.people = ["Tasib", "Rashed", "Sajib", "Emon", "Yousuf", "Wany", "Atif", "Akib"]

        # Create input fields dictionaries
        self.meal_entries = {}
        self.deposit_entries = {}

        self.create_widgets()

    def create_widgets(self):
        """Create all GUI widgets."""
        # Header
        header_frame = tk.Frame(self.root, bg=self.header_color, height=60)
        header_frame.pack(fill=tk.X)

        header_label = tk.Label(
            header_frame, 
            text="üçΩÔ∏è Meal Cost Calculator", 
            font=("Arial", 20, "bold"),
            bg=self.header_color,
            fg="white"
        )
        header_label.pack(pady=15)

        # Main container
        main_frame = tk.Frame(self.root, bg=self.bg_color)
        main_frame.pack(fill=tk.BOTH, expand=True, padx=20, pady=20)

        # Input section
        input_frame = tk.LabelFrame(
            main_frame, 
            text="Enter Data", 
            font=("Arial", 12, "bold"),
            bg=self.frame_color,
            padx=15,
            pady=15
        )
        input_frame.pack(fill=tk.BOTH, expand=True)

        # Create table header
        tk.Label(input_frame, text="Name", font=("Arial", 10, "bold"), 
                bg=self.frame_color, width=15).grid(row=0, column=0, padx=5, pady=5)
        tk.Label(input_frame, text="Meals", font=("Arial", 10, "bold"), 
                bg=self.frame_color, width=10).grid(row=0, column=1, padx=5, pady=5)
        tk.Label(input_frame, text="Deposit (Tk)", font=("Arial", 10, "bold"), 
                bg=self.frame_color, width=15).grid(row=0, column=2, padx=5, pady=5)

        # Create input fields for each person
        for i, person in enumerate(self.people, start=1):
            # Name label
            tk.Label(
                input_frame, 
                text=person, 
                font=("Arial", 10),
                bg=self.frame_color,
                anchor="w"
            ).grid(row=i, column=0, padx=5, pady=5, sticky="w")

            # Meals entry
            meal_entry = tk.Entry(input_frame, font=("Arial", 10), width=12)
            meal_entry.grid(row=i, column=1, padx=5, pady=5)
            meal_entry.insert(0, "0")
            self.meal_entries[person] = meal_entry

            # Deposit entry
            deposit_entry = tk.Entry(input_frame, font=("Arial", 10), width=17)
            deposit_entry.grid(row=i, column=2, padx=5, pady=5)
            deposit_entry.insert(0, "0")
            self.deposit_entries[person] = deposit_entry

        # Buttons frame
        button_frame = tk.Frame(main_frame, bg=self.bg_color)
        button_frame.pack(pady=15)

        # Calculate button
        calculate_btn = tk.Button(
            button_frame,
            text="Calculate",
            font=("Arial", 12, "bold"),
            bg=self.button_color,
            fg="white",
            width=15,
            command=self.calculate,
            cursor="hand2"
        )
        calculate_btn.grid(row=0, column=0, padx=10)

        # Reset button
        reset_btn = tk.Button(
            button_frame,
            text="Reset",
            font=("Arial", 12, "bold"),
            bg="#f44336",
            fg="white",
            width=15,
            command=self.reset,
            cursor="hand2"
        )
        reset_btn.grid(row=0, column=1, padx=10)

        # Results frame
        self.results_frame = tk.LabelFrame(
            main_frame,
            text="Results",
            font=("Arial", 12, "bold"),
            bg=self.frame_color,
            padx=15,
            pady=15
        )
        self.results_frame.pack(fill=tk.BOTH, expand=True, pady=(10, 0))

        self.results_text = tk.Text(
            self.results_frame,
            font=("Arial", 10),
            height=10,
            wrap=tk.WORD,
            bg="#f9f9f9",
            relief=tk.FLAT
        )
        self.results_text.pack(fill=tk.BOTH, expand=True)

    def calculate(self):
        """Calculate meal costs and display results."""
        try:
            # Collect data
            data = {}
            for person in self.people:
                meals = int(self.meal_entries[person].get())
                deposit = int(self.deposit_entries[person].get())
                data[person] = {"meals": meals, "deposit": deposit}

            # Calculate totals
            total_meals = sum(d["meals"] for d in data.values())
            total_deposit = sum(d["deposit"] for d in data.values())

            if total_meals == 0:
                messagebox.showerror("Error", "Total meals cannot be zero!")
                return

            # Calculate meal rate
            meal_rate = total_deposit / total_meals

            # Clear previous results
            self.results_text.delete(1.0, tk.END)

            # Display summary
            self.results_text.insert(tk.END, "=" * 60 + "\n")
            self.results_text.insert(tk.END, "SUMMARY\n")
            self.results_text.insert(tk.END, "=" * 60 + "\n")
            self.results_text.insert(tk.END, f"Total Meals: {total_meals}\n")
            self.results_text.insert(tk.END, f"Total Deposit: Tk {total_deposit}\n")
            self.results_text.insert(tk.END, f"Meal Rate: Tk {meal_rate:.2f} per meal\n\n")

            # Display individual results
            self.results_text.insert(tk.END, "=" * 60 + "\n")
            self.results_text.insert(tk.END, "INDIVIDUAL RESULTS\n")
            self.results_text.insert(tk.END, "=" * 60 + "\n")

            for person, person_data in data.items():
                balance = person_data["deposit"] - (person_data["meals"] * meal_rate)
                if balance > 0:
                    self.results_text.insert(tk.END, f"‚úì {person} will collect: Tk {balance:.2f}\n")
                else:
                    self.results_text.insert(tk.END, f"‚úó {person} has to deposit: Tk {abs(balance):.2f}\n")

        except ValueError:
            messagebox.showerror("Error", "Please enter valid numbers!")

    def reset(self):
        """Reset all input fields and results."""
        for person in self.people:
            self.meal_entries[person].delete(0, tk.END)
            self.meal_entries[person].insert(0, "0")
            self.deposit_entries[person].delete(0, tk.END)
            self.deposit_entries[person].insert(0, "0")

        self.results_text.delete(1.0, tk.END)

def main():
    root = tk.Tk()
    app = MealCalculatorGUI(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
